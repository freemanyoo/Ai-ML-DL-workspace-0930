```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
# # 1 Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞ (ÎãπÎá®Î≥ë Îç∞Ïù¥ÌÑ∞ÏÖã)
# from sklearn.datasets import load_diabetes
#
# diabetes = load_diabetes()
# df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
# df['target'] = diabetes.target
# print("--- Ï†ÄÏû•Ìï† Îç∞Ïù¥ÌÑ∞ (DataFrame) ---")
# print(df.head())
# print("---------------------------------")
# df.to_csv('diabetes.data.csv', index=False, encoding='utf-8-sig')
#
# print("\n'diabetes.data.csv' ÌååÏùºÏù¥ ÏÑ±Í≥µÏ†ÅÏúºÎ°ú Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§.")
```

```python
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()

X = diabetes.data
y = diabetes.target.reshape(-1, 1)  # ÎùºÎ≤® Îç∞Ïù¥ÌÑ∞ (0, 1, 2)

# 2 Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨ (Ï†ïÍ∑úÌôî Î∞è ÌõàÎ†®/ÌÖåÏä§Ìä∏ Î∂ÑÌï†)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# random_state=42, ÎèôÏùºÏÑ± Î≥¥Ïû•, Í∞ôÏùÄ Îç∞Ïù¥ÌÑ∞Î•º ÎΩëÎäî Ïö©ÎèÑ
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 3 PyTorch ÌÖêÏÑú Î≥ÄÌôò
X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32)
```

```python
# 4 MLP Î™®Îç∏ Ï†ïÏùò
class Regressor(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(10, 50)  # ÏûÖÎ†•Ï∏µ ‚Üí ÏùÄÎãâÏ∏µ1
        self.fc2 = nn.Linear(50, 30)  # ÏùÄÎãâÏ∏µ1 ‚Üí ÏùÄÎãâÏ∏µ2
        self.fc3 = nn.Linear(30, 1)  # ÏùÄÎãâÏ∏µ2 ‚Üí Ï∂úÎ†•Ï∏µ
        self.dropout = nn.Dropout(0.5)  # 50% ÎìúÎ°≠ÏïÑÏõÉ
        self.relu = nn.ReLU()  # ÌôúÏÑ±Ìôî Ìï®Ïàò

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.dropout(x)  # ÎìúÎ°≠ÏïÑÏõÉ Ï†ÅÏö©
        x = self.relu(self.fc2(x))
        x = self.fc3(x)  # Ï∂úÎ†•Ï∏µ (ÌöåÍ∑ÄÏù¥ÎØÄÎ°ú ÌôúÏÑ±Ìôî Ìï®Ïàò ÏóÜÏùå)
        return x
```

```python
# 5 Î™®Îç∏, ÏÜêÏã§ Ìï®Ïàò, ÏµúÏ†ÅÌôî Ìï®Ïàò ÏÑ§Ï†ï
model = Regressor()
criterion = nn.MSELoss()  # ÏÜêÏã§ Ìï®Ïàò (MSE)
optimizer = optim.Adam(model.parameters(), lr=0.001)  # Adam ÏµúÏ†ÅÌôî

# 6 ÌïôÏäµ ÏÑ§Ï†ï
num_epochs = 1000
batch_size = 32
train_losses = []

# 7 Î∞∞Ïπò ÌïôÏäµÏùÑ ÏúÑÌïú Îç∞Ïù¥ÌÑ∞Î°úÎçî ÏÉùÏÑ±
train_dataset = torch.utils.data.TensorDataset(X_train, y_train)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# 8 Î™®Îç∏ ÌïôÏäµ
for epoch in range(num_epochs):
    epoch_loss = 0.0
    for batch_X, batch_y in train_loader:
        optimizer.zero_grad()
        predictions = model(batch_X)
        loss = criterion(predictions, batch_y)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    # ÌèâÍ∑† ÏÜêÏã§ Ï†ÄÏû•
    train_losses.append(epoch_loss / len(train_loader))

    if (epoch + 1) % 100 == 0:
        print(f"Epoch [{epoch + 1}/{num_epochs}], Loss: {epoch_loss:.4f}")

# 9 ÏÜêÏã§Í∞í Í∑∏ÎûòÌîÑ Ï∂úÎ†•
plt.plot(train_losses)
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.title('Training Loss Over Epochs')
plt.show()
```

```python
# üîü Î™®Îç∏ ÌèâÍ∞Ä Î∞è ÏòàÏ∏°
model.eval()
with torch.no_grad():
    y_pred = model(X_test).numpy()

# üîü ÏòàÏ∏°Í∞í vs Ïã§Ï†úÍ∞í ÏãúÍ∞ÅÌôî
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Actual Quality")
plt.ylabel("Predicted Quality")
plt.title("Wine Quality Prediction")
plt.show()

# üîü ÏÑ±Îä• ÌèâÍ∞Ä (MSE)
test_mse = mean_squared_error(y_test, y_pred)
print(f"Test MSE: {test_mse:.4f}")
```

```python
# ‚≠ê ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ 5Í∞úÎ•º Ï∂úÎ†•ÌïòÍ≥† Ïã§Ï†ú ÏßÑÌñâÎèÑ vs ÏòàÏ∏° ÏßÑÌñâÎèÑ ÎπÑÍµê ‚≠ê
def predict_multiple_samples(X_data, y_true, num_samples=5):
    """
    Îç∞Ïù¥ÌÑ∞ÏÖãÏóêÏÑú num_samplesÍ∞úÏùò ÏÉòÌîåÏùÑ Î¨¥ÏûëÏúÑÎ°ú ÏÑ†ÌÉùÌïòÏó¨
    Î™®Îç∏Ïùò ÏòàÏ∏°Í∞íÍ≥º Ïã§Ï†ú Í∞íÏùÑ ÎπÑÍµê Ï∂úÎ†•Ìï©ÎãàÎã§.
    """
    model.eval()
    # Îç∞Ïù¥ÌÑ∞ÏÖãÏóêÏÑú num_samples Í∞úÏàòÎßåÌÅº Î¨¥ÏûëÏúÑÎ°ú Ïù∏Îç±Ïä§ ÏÑ†ÌÉù
    indices = np.random.choice(len(X_data), num_samples, replace=False)
    samples = X_data[indices]
    true_values = y_true[indices].reshape(-1)  # Ïã§Ï†ú ÏßÑÌñâÎèÑ Í∞í

    # ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ Ï†ïÍ∑úÌôî
    samples_scaled = scaler.transform(samples)
    samples_tensor = torch.tensor(samples_scaled, dtype=torch.float32)

    # Î™®Îç∏ ÏòàÏ∏° ÏàòÌñâ
    with torch.no_grad():
        predicted_values = model(samples_tensor).numpy().flatten()

    # üö® 1~9 Îì±Í∏âÏúºÎ°ú Î∞îÍæ∏Îäî Î∂àÌïÑÏöîÌïú Î°úÏßÅÏùÑ ÏôÑÏ†ÑÌûà ÏÇ≠Ï†úÌñàÏäµÎãàÎã§.

    # Í≤∞Í≥º Ï∂úÎ†• (Ïö©Ïñ¥Î•º 'ÏôÄÏù∏ ÌíàÏßà'ÏóêÏÑú 'ÎãπÎá®Î≥ë ÏßÑÌñâÎèÑ'Î°ú Î™®Îëê ÏàòÏ†ï)
    print("\nüîç ÏÉòÌîå 5Í∞ú Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏° Í≤∞Í≥º:")
    print("=" * 50)
    print(f"{'ÏÉòÌîå Ïù∏Îç±Ïä§':<12}{'Ïã§Ï†ú ÏßÑÌñâÎèÑ':<15}{'ÏòàÏ∏° ÏßÑÌñâÎèÑ'}")
    print("=" * 50)
    for i, idx in enumerate(indices):
        print(f"{idx:<12}{true_values[i]:<15.2f}{predicted_values[i]:.2f}")


# üèÜ ÏÉòÌîå 5Í∞ú ÏòàÏ∏° Ïã§Ìñâ
# ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ XÏôÄ yÎ•º ÏÇ¨Ïö©Ìï©ÎãàÎã§.
predict_multiple_samples(X, y)
```

```python

# ‚≠ê ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ 5Í∞úÎ•º Ï∂úÎ†•ÌïòÍ≥† Í∏∞Ï°¥ ÌíàÏßà vs ÏòàÏ∏°Îêú ÌíàÏßà ÎπÑÍµê ‚≠ê
def predict_multiple_samples(X_data, y_true, num_samples=5):
    """
    Îç∞Ïù¥ÌÑ∞ÏÖãÏóêÏÑú num_samplesÍ∞úÏùò ÏÉòÌîåÏùÑ ÏÑ†ÌÉùÌïòÏó¨ Î™®Îç∏ ÏòàÏ∏°Í∞íÍ≥º Ïã§Ï†ú ÌíàÏßàÏùÑ ÎπÑÍµê Ï∂úÎ†•
    """
    model.eval()
    indices = np.random.choice(len(X_data), num_samples, replace=False)  # Î¨¥ÏûëÏúÑ 5Í∞ú ÏÉòÌîå ÏÑ†ÌÉù
    samples = X_data[indices]  # ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ ÏÑ†ÌÉù
    true_quality = y_true[indices].reshape(-1)  # Ïã§Ï†ú ÌíàÏßà

    # ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ Ï†ïÍ∑úÌôî
    samples_scaled = scaler.transform(samples)
    samples_tensor = torch.tensor(samples_scaled, dtype=torch.float32)

    # Î™®Îç∏ ÏòàÏ∏° ÏàòÌñâ
    with torch.no_grad():
        predicted_quality = model(samples_tensor).numpy().flatten()

    # üîπ 1~9 Î≤îÏúÑ ÎÇ¥Î°ú Î≥ÄÌôò (Î∞òÏò¨Î¶º ÌõÑ Î≤îÏúÑ Ï†úÌïú)
    predicted_quality_int = [min(9, max(1, round(q))) for q in predicted_quality]

    # Í≤∞Í≥º Ï∂úÎ†•
    print("\nüîç ÏÉòÌîå 5Í∞ú Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏° Í≤∞Í≥º:")
    print("=" * 60)
    print(f"{'ÏÉòÌîå Ïù∏Îç±Ïä§':<12}{'Ïã§Ï†ú ÌíàÏßà':<12}{'ÏòàÏ∏° ÌíàÏßà (Ïã§Ïàò)':<20}{'ÏòàÏ∏° ÌíàÏßà (Ï†ïÏàò)'}")
    print("=" * 60)
    for i, idx in enumerate(indices):
        print(f"{idx:<12}{true_quality[i]:<12.0f}{predicted_quality[i]:<20.2f}{predicted_quality_int[i]}")


# üèÜ ÏÉòÌîå 5Í∞ú ÏòàÏ∏° Ïã§Ìñâ
predict_multiple_samples(X, y)
```
